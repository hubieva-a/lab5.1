```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
#В тот момент, когда мимо станции со скоростью 5 м/с проходил товарный состав, 
#от платформы в том же направлении отошёл пассажирский поезд. Через какое время 
#пассажирский поезд догнал товарный, если пассажирский двигался с ускорением 0,3 м/с2, 
#а товарный – равномерно?

t = [0, 10, 20, 30, 40]
x1 = [i*5 for i in t]
x2 = [0.15*i*i for i in t]

plt.title("The graph for the first train is red, the second graph is blue")
plt.xlabel("Time, t")
plt.ylabel("Distance, x1, x2")
plt.grid(True)
plt.plot(t, x1, '--r', t, x2, 'b')
```

```python
#Дан следующий вариационный ряд
#1 2 3 4 5 6 7 7 6 4
#1 1 2 2 4 4 4 5 5 5
#Требуется Построить полигон распределения

new_list = [1, 2, 3, 4, 5, 6, 7, 7, 6, 4, 1, 1, 2, 2, 4, 4, 4, 5, 5, 5,]
numbers = [1, 2, 3, 4, 5, 6, 7,]
k1=0
k2=0
k3=0
k4=0
k5=0
k6=0
k7=0
for i in new_list:
    if i == 1:
        k1+=1
    elif i == 2:
        k2+=1
    elif i == 3:
        k3+=1
    elif i == 4:
        k4+=1
    elif i == 5:
        k5+=1
    elif i == 6:
        k6+=1
    elif i == 7:
        k7+=1

value_list = [k1, k2, k3, k4, k5, k6, k7]
plt.bar(numbers, value_list)
plt.title("A diagram")
```

```python
#В среднем на одного взрослого трудоспособного человека из продуктов 
#питания в год полагается 18,5 кг рыбы, 58,6 кг мяса, 60 кг фруктов, 
#114,6 кг овощей, 126,5 кг хлебных продуктов. Построить круговую диаграмму распределения 
#типов пищевой продукции, потребляемой человеком

values = [18.5, 58.6, 60, 114.6, 126.5]
labels = ["fish", "meat", "fruit", "vegetables", "grain"]
fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.1f%%')
ax.axis("equal")
```
